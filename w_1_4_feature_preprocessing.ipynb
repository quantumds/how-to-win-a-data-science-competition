2. Feature Generation

Se dividen los datos en Train, Test y Validation

Existe una relacion lineal entre la variable respuesta (numero de manzanas vendidas) y el tiempo que avnza. es un ejercicio de fporecasting

**la generacion de variables para que el modelo entienda esta relacin debe depender tambien del modelo q estas usando. Si usas un modelo lineal simplemente aadiendo el week number sera suficiente para q el modelo lo entienda. (los dias son indices en un forecasting y no se usan como predictor). Por otro lado, si usas un arbol, debes hacer un join con una tabla maestra que tenga el numero de manzanas medios por dia, o el stddev de manzanas por dia, para que cuando el arbol haga cortes, ve que al numero de dias siempre le toca el mismo avg, y que este avg crece si los dias crecen.**

modelo lineal usa numero de la semana
modelo random forest usa avg manzanas vendidas por dia y hace una join con estos valores maestros
