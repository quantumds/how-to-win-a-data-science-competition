There are different types of categorical features:

1) Nominal features: They cannot be organized: Favourite artist, Continets, etc.<br>
2) Ordinal features: They can be organized: Militar rank, grades, etc.<br>
3) Identifiers

For linear models: one hot encoding <br>
For tree-based models: label encoding<br>
For both models: frequency encoding
