Pre-processing in numeric features should be divided in:
    
    1) Tree based models
    2) Non tree based models
    

Non tree based models require **scaling**

We will explain why

there is a classification metric that is called l2 distance (linear model) that assigns if a point is from one label or other (response variable) based to the distance to other labeled points

If we do not regularize in this type of problem, the scale of the variable will play a huge impact. imagine that variable 2 is price of the house. this values comes in millions, and the other variable is number of kids, these variable comes in units. Distance of our point that we want to predict if it is class a or b will always classify class a for any point, because distance of the point in the variable 1 space (apartment house) is much bigger than distance in variable 2 space (number of kids). 

it is what we see here:

```python
from IPython.display import Image
Image(filename="C:/ANIBAL/FORMATION/PERSONAL/how_to_win_a_data_science_competition/images/num1.PNG")
```

the third case is extreme: algorithm will not even have into account the x2 variable (number of kids) because if we multiply the apartment price by 0, all possible points will always be of X mark class instead of O (the X are blue and = are read, ? is the point we need to asign based on distance). If we multiply by hundreds of millions then it will be always O type instead of X type.

Conclusion this type of non tree algorithms require scaling so that the scale or unit of the variables doesnt impact results and all play in the same league. 

Gradient descent methods can go crazy without proper scaling.
